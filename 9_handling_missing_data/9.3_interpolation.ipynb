```python
import pandas as pd 
```

```python
# sample data frame:
Employee_data = {
    "name": ["Bilal Kiani", "Afan", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, None, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, None, 40000, 50000, None, 60000, 43000, 50500],
    "performance_score": [50, None, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# Way-3    interpolation,  this is a technique by which we fill and estimated value in place of missing values, e.g.  we have a list[10, 20,  , 40, 50] in this list one value is missing but we can estimate it by series what will be value here this is an interpolation. only for numerical columns
# interpolation are like:   linear interpolation, polynomial interpolation, time,   these are important but there are more inpandas

#why to use interpolation: 
#     to preserve data integritiy
#     smooth trends 
#     avoid data loss 
# for this pandas provides us a method  interpolate()
# syntex:   df.interpolate(method, axis= 0 or 1,  inplace= True)

df.interpolate(method="linear", axis=0, inplace=True)
print(df)
```

```python
# we use interpolation when we are working with:
#      time series data, 
#      numeric data with trends, 
#      avoid droping rows 
# disadvantegs: 
#     it does not works with catagorical data, 
```

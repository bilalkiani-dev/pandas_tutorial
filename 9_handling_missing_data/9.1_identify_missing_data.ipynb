```python
import pandas as pd 
```

```python
# representation of missing values:     1- NaN (not a number),    2- None (for object data types)
# first we hwave to identify missing values,   for this pandas provide us a method   isnull()
# when using isnull()  method, it wil return boolien dataframe, as True or False,  if True means NaN is missing, value is missing, if False means value is present not missing

# let say we have this sample data frame, containing whole row two none value:
Employee_data = {
    "name": ["Bilal Kiani", None, "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, None, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, None, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, None, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# now we have to identify we use .isnull()
print(df.isnull())
```

```python
# now if we have to know number of missing values we use .isnull().sum
print(df.isnull().sum())
```

```python
import pandas as pd 
```

```python
# consider we have a dataset containing 1000 rows,  and checking whole data set may be impractical, so we can explore some rows, like first 5 rows or last some rows etc 
#for this pandas provides us functions   "head()"  "tail()"
#syntex for head:   head(n)  :    this will access first n rows,    here n means number of rows, if we don't pass n, by default it will show 5 first rows
#syntex for tail:   tail(n)  :    this will access last n rows,    here n means number of rows, if we don't pass n, by default it will show 5 last rows
```

```python
df = pd.read_excel("StudentsPerformance.xlsx")
print("print first 10 rows: \n", df.head(10))
```

```python
print("Last 10 rows: \n", df.tail(10))
```

```python
#when we dont pass "n" then prints only 5 rows    first or last
print(df.head())
print(df.tail())
```

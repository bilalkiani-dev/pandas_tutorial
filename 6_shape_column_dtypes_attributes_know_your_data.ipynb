```python
import pandas as pd 
```

```python
Employee_data = {
    "name": ["Bilal Kiani", "Afan Hafiz", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, 23, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, 30000, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, 60, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# first we have to know about our data,   
# how big is our data, no of rows and columns, this helps to understand the scale of dataset 
# name of column of data set, so that we can access them and can modify as required 

#to solve these problems pandas provides us three attributes, 'shape', 'column' and 'dtype'
#shape:  returns a tuple containing two values   no of rows and no of column 
# column:  shows the names of all column in dataset 
# dtype will show the data type of that column values

print(f'Shape: {df.shape}')
print(f'Column_name: {df.columns}')
print(df.shape)        # Shows (Rows, Columns) -> (5, 4)
print(df.dtypes)       # Shows data type of each column (Object means Text)
```

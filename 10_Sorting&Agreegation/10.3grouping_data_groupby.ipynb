```python
import pandas as pd 
```

```python
# Sample Data
Employees_data = {
    "Name": ["Bilal", "Afan", "Saqib", "Junaid", "Ali", "Sara", "Asad", "Asghar", "Mamoon"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 30000, 45000, 60000, 32000, 52000, 55000, 40000, 48000],
    "Experience_Years": [2, 1, 2, 5, 2, 3, 6, 3, 4]
}

df = pd.DataFrame(Employees_data)
print("Original Data:")
print(df)
```

```python
# .groupby  method:  when we need to group to column and then perform statistical operation we use .groupby   methhod, like
# in above data set we have columns  "Department" and "Salary",  i wana check that hwo much sallary goes to each department, means sum of each department salary,
# syntex:    df.groupby("column1")["column2"].statistical_operation
each_Dpt_Total_sal = df.groupby("Department")["Salary"].sum()
print("Total Sallaries goes to Each Department are: ")
print(each_Dpt_Total_sal)
```

```python
# now i wana combine columns and then find statistical operation 
# syntex:   df.groupby(["column1", "column2", ...]).["column3"].sts_opr     
sal_name_dep = df.groupby(["Department", "Name"])["Salary"].sum()   # normally this is ued when we wana calculate something based on multiple column, 
print(sal_name_dep)
```

```python
# and if wana find min, max, sum etc at once we will use .agg() method
# syntex:    df.groupby("column_-name")["column_name"].agg(['min', 'max', 'mean', ...])
department_sts = df.groupby("Department")["Salary"].agg(['min', 'max', 'mean'])
print("Department statistics: ")
print(department_sts)
```

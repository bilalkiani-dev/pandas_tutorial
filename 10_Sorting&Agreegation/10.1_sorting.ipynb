```python
# sorting is the meann of arranging the data based on the values of one or more columns:
# we cn sort data in ascending or decendingn order both alphatically or numerically 
# for this we use a function    sort_values()     
import pandas as pd 
```

```python
# sample data frame
Employee_data = {
    "name": ["Bilal Kiani", "Afan Hafiz", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, 23, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, 30000, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, 60, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
#Sorting data one column: 
# syntex: df.sort_values(by="column_name", ascendin or decending , inplace)

df.sort_values(by="name", ascending=True, inplace=True)  # if i do ascending = false   then data will sort in decending order
df.reset_index(drop=True, inplace=True) #.reset_index() used when we arrange the data the number at left sides will also be moved along arranged data, To make it look clean (0, 1, 2, 3...), use .reset_index().
print("Data Sorted in Ascending order based on column name: ")
print(df)
```

```python
# if we wana sort based on multiple columns: 
# This tells Pandas the order of priority:
# Priority 1: First, sort everyone by their performance_score.
# Priority 2 (The Tie-Breaker): If (and only if) two employees have the exact same performance score, then look at their sallary to decide who comes first.
# syntex:    df.sort_values(by=["column1", "column2", ...], ascending = [True/False, True/False, ...], inplace = True/False)

df.sort_values(by=["performance_score", "sallary"], ascending=[False, False], inplace=True)
df.reset_index(drop=True, inplace=True)   # by using .reset_index()   the indexs will remain clean and consistant as 0,1,2,3,...
print("Data sorted who have more score will get more sallary: ")
print(df)
```

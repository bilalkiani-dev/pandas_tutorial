```python
# Aggregation is different. It takes many rows and squashes them down into a single number (like a sum, average, or count).
import pandas as pd 
```

```python
# Sample Data
Employees_data = {
    "Name": ["Bilal", "Afan", "Saqib", "Junaid", "Ali", "Sara", "Asad", "Asghar", "Mamoon"],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 30000, 45000, 60000, 32000, 52000, 55000, 40000, 48000],
    "Experience_Years": [2, 1, 2, 5, 2, 3, 6, 3, 4]
}

df = pd.DataFrame(Employees_data)
print("Original Data:")
print(df)
```

```python
# if i wana check the average sallary we calculate mean()
average_sallary = df["Salary"].mean()  # this will give us the average sallary 
print(average_sallary)
# same i wana check maximum and minimum year_experience, and also wana check the sum of sallary 
max_exp = df["Experience_Years"].max()
min_exp = df["Experience_Years"].min()
total_sal = df["Salary"].sum()
print("Maximum Experience is: ", max_exp, "years and minimum Experience is: ", min_exp, "years and the total salary goes to Employees is: ", total_sal)
```

```python
# 1. The .map() Method (The Simple Swapper)
# Use Case: You have a code (like "M" or "F") and you want to swap it for a full name ("Male" or "Female"). It works exactly like looking up words in a dictionary.

# Example: changing Department codes to full names.

import pandas as pd

df = pd.DataFrame({
    "Name": ["Bilal", "Sara", "Ali"],
    "Dept_Code": ["IT", "HR", "IT"]
})

# 1. Create your Dictionary ( The "Map" )
#    Key = Old Value, Value = New Value
dept_dictionary = {
    "IT": "Information Technology",
    "HR": "Human Resources"
}

# 2. Use .map() to swap the values
df["Department"] = df["Dept_Code"].map(dept_dictionary)

print(df)
# What happened? Pandas looked at "IT", found it in the dictionary, and replaced it. Simple.
```

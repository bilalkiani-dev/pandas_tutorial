```python
import pandas as pd

data = {
    "Employee_ID": ["EMP-101", "EMP-102", "EMP-103", "EMP-104"],
    "Name": ["Bilal Kiani", "Afan Hafiz", "Saqib Sheikh", "Junaid Khan"],
    "Salary": [50000, 70000, 45000, 60000],
    "Bonus": [5000, 7000, 4500, 6000],
    "Department": ["IT Support", "HR Manager", "IT Developer", "Finance Lead"]
}

df = pd.DataFrame(data)

print("--- Original Data ---")
print(df)
```

```python
# Part 1: Column Math (Creating New Columns)
# This works just like Excel formulas. You select column A, select column B, and add them.
# Syntax: df["New_Column"] = df["A"] + df["B"]
# 1. Basic Math (+, -, *, /) 

# "Calculate Total Income (Salary + Bonus)."
# Create a new column 'Total_Income'
df["Total_Income"] = df["Salary"] + df["Bonus"]
print("\n--- Added Total Income ---")
print(df[["Name", "Salary", "Bonus", "Total_Income"]])
```

```python
# 2. Constant Math
# "Give everyone a 20% raise."

# Multiply the whole column by 1.20
df["New_Salary"] = df["Salary"] * 2.10
print(df)
```

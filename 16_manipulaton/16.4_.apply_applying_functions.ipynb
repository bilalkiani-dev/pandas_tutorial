```python
# 2. The .apply() Method (The Custom Logic)
# Use Case: You need to do math or logic that isn't just a simple swap. Maybe the result depends on the number being high or low.
# Example: We want to give a Rating based on the Salary.
# If Salary > 50,000 → Rating is "High"
# If Salary <= 50,000 → Rating is "Low"
# You cannot do this with a simple dictionary. You need a Function.

import pandas as pd

df = pd.DataFrame({
    "Name": ["Bilal", "Sara", "Ali"],
    "Salary": [60000, 45000, 80000]
})

# 1. Define a Python Function (The Logic)
#    This function takes ONE number (x) and returns a result
def get_rating(x):
    if x > 50000:
        return "High Earner"
    else:
        return "Standard"

# 2. Apply this function to every single row in the column
#    Pandas takes 60000, puts it into 'x', runs the code, and saves the result.
df["Rating"] = df["Salary"].apply(get_rating)

print(df)

# What happened? Pandas took every salary one by one, sent it to your get_rating function, and stored the answer it got back.
```

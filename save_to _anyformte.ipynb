```python
import pandas as pd 
#creating dictionary
students = {
    "name": ["Bilal", "Afan", "Shabbar", "Saood"],
    "RollNo": [103, 68, 105, 71],
    "CGPA": [3.21, 3.00, 2.5, 3.4]
}
print(students)
```

```python
df = pd.DataFrame(students)
print(df)
```

```python
#we are saving this dictionary to csv file , also we use index= false to not write theh index, as in above cell we can sea indexes
df.to_csv("students_data.csv", index=False)
```

```python
#now save students dictionary as xlsx file 
df.to_excel("students_data.xlsx", index=False)
```

```python
#now save this dictionary to json file 
df.to_json("students_data.json", index=False)
```

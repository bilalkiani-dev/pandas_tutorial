```python
import pandas as pd 
```

```python
# sample data frame
Employee_data = {
    "name": ["Bilal Kiani", "Afan Hafiz", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, 23, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, 30000, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, 60, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# there are two methods of adding new column in dataset 
# 1- adding through assignment, and it will add column at the end of all columns
#syntex:       df["column_name"] = some_data
print("Employee data now with extra column Bonous: ")
df["Bonous"] = df["sallary"]*0.1    # by this we added new column "bonous" at last,   bonous is 10 percent of sallary
print(df)
```

```python
# 2- adding using insert()  method,  using this method we can add column at specific location where we want 
# syntex:      df.insert(location_value, "column_name", some_data)
print("Employee data now having employee id: ")
df.insert(0, "EmployeeID", [10,20,30,40,50,60,70,80])
print(df)
```

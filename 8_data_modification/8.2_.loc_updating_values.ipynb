```python
import pandas as pd 
```

```python
# sample data frame
Employee_data = {
    "name": ["Bilal Kiani", "Afan Hafiz", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, 23, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, 30000, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, 60, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# two ways of updating values: 
# 1- updating a specific cell, for this we use a method   .loc[]
# syntex: df.loc[row_index, "column_name"] = new_value
print("Sallary of Ameen Khan is inreased 4500: ")
df.loc[7, "sallary"] = 55000
print(df)
```

```python
#2- updating the values of whole column: 
#syntex:   df["column_name"] = new data ,  simple condition 
print("Increasing sallaries of Employees 5%: ")
df["sallary"] = df["sallary"]*1.05
print(df)
```

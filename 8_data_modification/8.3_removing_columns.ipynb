```python
import pandas as pd 
```

```python
# sample data frame
Employee_data = {
    "name": ["Bilal Kiani", "Afan Hafiz", "Shabbar Alam", "Saood Sheikh", "Junaid Kiani", "Abdul Qadar", "Saqib Sheikh", "Ameen Khan"],
    "age": [22, 23, 24, 21, 43, 44, 34, 50],
    "sallary": [20000, 30000, 40000, 50000, 45000, 60000, 43000, 50500],
    "performance_score": [50, 60, 89, 78, 68, 90, 95, 86]
}
df = pd.DataFrame(Employee_data)
print("Sample DataFrame: ")
print(df)
```

```python
# for removing any column pandas provides a method  .drop(),  
# syntex:   df.drop(columns= ["column_name"], inplace = true or false), 
# here inplace means when it is true the our modification will be in original dataset,
# but if inplace = flase means it will create new data set, will not updata the original one dataset

print("Modified Dataset: ")
df.drop(columns=['performance_score'], inplace=True) # here we deleted a column "performance_score"
print(df)
```

```python
# and if we need to delete multiple columns at a time then syntex is:      df.drop(columns= ["column1", "column2", ...], inplace = True)
print("Modified Dataset: ")
df.drop(columns=["age", "sallary"], inplace=True)   # here i deleted two more columns, age and sallary
print(df)
```
